```python
from PIL import Image
import matplotlib.pyplot as plt
import PIL
import cv2
import numpy as np
```

```python
# 读取两张图像
image1 = Image.open('images/222.png').convert("L")
image2 = Image.open('2without2to222.png').convert("L")

# 确保两张图像具有相同的尺寸
if image1.size != image2.size:
    raise ValueError("图像尺寸不匹配")

# 将图像转换为NumPy数组
array1 = np.array(image1)
array2 = np.array(image2)

# 计算两个数组的差值
result_array = array1 - array2

# 将结果数组转换回图像
result_image = Image.fromarray(result_array)

# 使用Matplotlib显示结果图像
plt.imshow(result_image)
plt.title('Image Subtraction Result')
plt.axis('off')
plt.show()
```
